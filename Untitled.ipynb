```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries loaded successfully!")
```

```python
df = pd.read_csv("data/Walmart.csv")
print("Dataset loaded!")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
print(df['Date'].head())
print(df['Date'].dtype)
```

```python
print("Earliest date:", df['Date'].min())
print("Latest date:", df['Date'].max())
print("Number of unique stores:", df['Store'].nunique())
```

```python
weekly_total = df.groupby('Date')['Weekly_Sales'].sum()

plt.figure(figsize=(14, 5))
plt.plot(weekly_total.index, weekly_total.values, color='teal')
plt.title('Total Weekly Sales Across All Stores (2010-2012)')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week

print(df[['Date', 'Year', 'Month', 'Week']].head())
```

```python
df['Days_Since_Start'] = (df['Date'] - df['Date'].min()).dt.days

print(df[['Date', 'Days_Since_Start']].head())
```

```python
df_model = df.drop(columns=['Date'])
print(df_model.columns.tolist())
```

```python
X = df_model.drop(columns=['Weekly_Sales'])
y = df_model['Weekly_Sales']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
split_index = int(len(df_model) * 0.8)

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

print("Training rows:", X_train.shape[0])
print("Testing rows:", X_test.shape[0])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling done!")
```

```python
df_sorted = df.sort_values(['Store', 'Date'])
df_sorted['Moving_Avg_Forecast'] = df_sorted.groupby('Store')['Weekly_Sales'].transform(lambda x: x.shift(1).rolling(window=4).mean())

print(df_sorted[['Store', 'Date', 'Weekly_Sales', 'Moving_Avg_Forecast']].head(10))
```

```python
from sklearn.metrics import mean_absolute_error

df_baseline = df_sorted.dropna(subset=['Moving_Avg_Forecast'])

baseline_mae = mean_absolute_error(df_baseline['Weekly_Sales'], df_baseline['Moving_Avg_Forecast'])
print("Baseline MAE:", round(baseline_mae, 2))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)

print("Model trained successfully!")
```

```python
y_pred = rf_model.predict(X_test_scaled)

ml_mae = mean_absolute_error(y_test, y_pred)
print("Random Forest MAE:", round(ml_mae, 2))
```

```python
df_sorted['Prev_Week_Sales'] = df_sorted.groupby('Store')['Weekly_Sales'].shift(1)
df_sorted['Avg_Last_4_Weeks'] = df_sorted.groupby('Store')['Weekly_Sales'].transform(lambda x: x.shift(1).rolling(window=4).mean())

print(df_sorted[['Store', 'Date', 'Weekly_Sales', 'Prev_Week_Sales', 'Avg_Last_4_Weeks']].head(10))
```

```python
df_features = df_sorted.dropna(subset=['Prev_Week_Sales', 'Avg_Last_4_Weeks'])

df_model2 = df_features.drop(columns=['Date', 'Moving_Avg_Forecast'])

print(df_model2.columns.tolist())
print("Rows remaining:", df_model2.shape[0])
```

```python
X2 = df_model2.drop(columns=['Weekly_Sales'])
y2 = df_model2['Weekly_Sales']

split_index2 = int(len(df_model2) * 0.8)

X2_train = X2.iloc[:split_index2]
X2_test = X2.iloc[split_index2:]
y2_train = y2.iloc[:split_index2]
y2_test = y2.iloc[split_index2:]

scaler2 = StandardScaler()
X2_train_scaled = scaler2.fit_transform(X2_train)
X2_test_scaled = scaler2.transform(X2_test)

print("Training rows:", X2_train.shape[0])
print("Testing rows:", X2_test.shape[0])
```

```python
rf_model2 = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model2.fit(X2_train_scaled, y2_train)

y_pred2 = rf_model2.predict(X2_test_scaled)

ml_mae2 = mean_absolute_error(y2_test, y_pred2)
print("Random Forest MAE (with lag features):", round(ml_mae2, 2))
```

```python
improvement = ((baseline_mae - ml_mae2) / baseline_mae) * 100
print("Improvement over baseline:", round(improvement, 2), "%")
```

```python
df_retail = pd.read_csv("data/recommendation/data.csv", encoding='ISO-8859-1')
print("Rows:", df_retail.shape[0])
print("Columns:", df_retail.shape[1])
df_retail.head()
```

```python
df_retail.isnull().sum()
```

```python
df_retail = df_retail.dropna(subset=['CustomerID', 'Description'])
print("Rows remaining:", df_retail.shape[0])
```

```python
df_retail = df_retail[~df_retail['InvoiceNo'].astype(str).str.startswith('C')]
print("Rows after removing cancellations:", df_retail.shape[0])
```

```python
print("Unique customers:", df_retail['CustomerID'].nunique())
print("Unique products:", df_retail['StockCode'].nunique())
```

```python
customer_product_matrix = df_retail.pivot_table(
    index='CustomerID', 
    columns='Description', 
    values='Quantity', 
    aggfunc='sum', 
    fill_value=0
)

print("Matrix shape:", customer_product_matrix.shape)
customer_product_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

product_similarity = cosine_similarity(customer_product_matrix.T)

product_similarity_df = pd.DataFrame(
    product_similarity, 
    index=customer_product_matrix.columns, 
    columns=customer_product_matrix.columns
)

print("Similarity matrix shape:", product_similarity_df.shape)
```

```python
def recommend_products(product_name, top_n=5):
    if product_name not in product_similarity_df.columns:
        return "Product not found in dataset"
    
    similar_scores = product_similarity_df[product_name].sort_values(ascending=False)
    similar_scores = similar_scores.drop(product_name)
    
    return similar_scores.head(top_n)

# Test it with a real product
recommend_products('WHITE HANGING HEART T-LIGHT HOLDER')
```

```python
recommend_products('RED WOOLLY HOTTIE WHITE HEART.')
```
